```python
#일반적인 자동차를 나타내는 클래스이다.

class Car:
  def __init__(self,make,model,color,price):
    self.make=make
    self.model=model
    self.color=color
    self.price=price

  def setMake(self,make): #설정자 메소드
    self.make=make

  def getMake(self):  #접근자 메소드
    return self.make
  
  #차량에 대한 정보를 문자열로 요약하여서 반환한다
  def getDesc(self):
    return "차량 =("+str(self.make)+","+\
      str(self.model)+","+\
      str(self.color)+","+\
      str(self.price)+")"

class ElectricCar(Car):
  def __init__(self,make,model,color,price,batterySize):
    super().__init__(make,model,color,price)
    self.batterySize=batterySize

  def setBatterySize(self, batterySize):  #설정자 메소드
      self.batterySize=batterySize

  def getBatterySize(self):   #접근자 메소드
      return self.batterySize

myCar = ElectricCar("Tsla","Model S","white",10000,0)
myCar.setMake("Tsla")  #설정자 메소드 호출
myCar.setBatterySize(60) #설정자 메소드 호출
print(myCar.getDesc()) #전기차 객체를 문자열로 호출
```

```python
class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  
  def show(self):
    print(self.name,self.age)

class Student:
  def __init__(self, id):
    self.id = id

  def getId(self):
    return self.id

class CollegeStudent(Person, Student):
  def __init__(self, name, age, id):
    Person.__init__(self, name, age)
    Student.__init__(self,id)

obj = CollegeStudent('Kim', 22, '100036')
obj.show()
print(obj.getId())
```

```python
#직원과 매니저

class Employee:
  def __init__(self,name,salary):
    self.name = name
    self.salary = salary

  def getSalary(self):
    return self.salary
  
  def __str__(self):
    return f"이름:{self.name}, 월급:{self.getSalary()}"

class Manager(Employee):
  def __init__(self, name, salary, bonus):
    super().__init__(name, salary) #부모 클래스 호출
    self.bonus = bonus
  
  def getSalary(self):
    return super().getSalary() + self.bonus #부모클래스의 getSalary()호출 + bonus

kim = Manager("김철수", 2000000, 1000000)
print(kim)

park = Employee("박박박", 1500000)
print(park)
```
